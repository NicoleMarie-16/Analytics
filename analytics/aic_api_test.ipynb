**Import Libraries**

```python
import httplib2
import json
import time
import pandas as pd
from pandas.io.json import json_normalize
from pprint import pprint
```

## API Exploration

**Establish Connection**

```python
h = httplib2.Http()
```

**Make the Calls**

```python
resp, content = h.request("https://api.artic.edu/api/v1/artworks?fields=id,title")
assert resp.status == 200
assert resp['content-type'] == 'application/json'
```

**Convert to JSON**

```python
content_str = content.decode('utf-8')
d = json.loads(content_str)
pprint(d['data'])
```

**How many records per page?**

```python
total_pages = d["pagination"]["total_pages"]

total_records = d["pagination"]["total"]

total_records/total_pages
```

## Loop Through Pages to Get All Results 

```python
def retry(fun, total_pages, max_tries=55):
    for i in range(max_tries):
        try:
           time.sleep(1.0) 
           fun(total_pages)
           break
        except Exception:
            continue
```

```python
def get_artworks(total_pages):
    results_list = []
    for i in range(1,total_pages+1):
        request_string = "https://api.artic.edu/api/v1/artworks?fields=id,title&page={page_num}".format(page_num=i)
        print("Retrieving results from page {page_num}".format(page_num=i))
        resp, content = h.request(request_string)
        content_str = content.decode('utf-8')
        d = json.loads(content_str)
        results_list.append(d)
    return results_list
```
